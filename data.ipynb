```python
import pandas as pd

df1 = pd.read_csv('rotten_tomatoes_critic_reviews.csv')
df2 = pd.read_csv("rotten_tomatoes_movies.csv")
```

### features

rotten_tomatoes_link: Link to the movie's page on the Rotten Tomatoes website.

critic_name: The name of the critic who reviewed the movie.

top_critic: Indicates whether the critic is classified as a top critic on the Rotten Tomatoes platform.

publisher_name: The name of the publisher or media outlet where the review was published.

review_type: Type of the review (e.g., fresh or rotten).

review_score: The score the critic gave to the movie, typically on a defined scale.

review_date: The date when the review was published.

review_content: The full text content of the review.

movie_title: The title of the movie that was reviewed.

movie_info: Information about the movie, such as a synopsis.

critics_consensus: A summary of the critical consensus for the movie.

content_rating: The age suitability rating of the movie (e.g., G, PG, PG-13, R, etc.).

genres: The genre or genres of the movie.

directors: The director or directors of the movie.

authors: The writer or writers of the movie.

actors: The main actor or actors in the movie.

original_release_date: The date when the movie was originally released.

streaming_release_date: The date when the movie was released on streaming platforms.

runtime: The duration of the movie.

production_company: The company or companies that produced the movie.

tomatometer_status: The status of the movie on the Tomatometer, which measures the percentage of approved critic reviews.

tomatometer_rating: The Tomatometer rating of the movie.

tomatometer_count: The number of critic reviews that make up the Tomatometer rating.

audience_status: The audience's general reception of the movie.

audience_rating: The average rating given by the audience.

audience_count: The number of audience members who rated the movie.

tomatometer_top_critics_count: The number of top critics whose reviews make up the Tomatometer rating.

tomatometer_fresh_critics_count: The number of critics who gave a "fresh" rating to the movie.

tomatometer_rotten_critics_count: The number of critics who gave a "rotten" rating to the movie.

```python
df = df1.merge(df2, how='left', on=list(set(df1.columns) & set(df2.columns)))

columns = ['critic_name', 'movie_title','movie_info','genres', 'review_content', 'review_type', 'review_score']
df = df[columns].dropna()

df.to_csv("data.csv", index=False)
from google.colab import files
files.download('data.csv')
```
